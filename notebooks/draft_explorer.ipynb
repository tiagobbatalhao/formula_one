```python
from pathlib import Path
import pandas as pd
```

```python
def find_files():
    path = Path('../data/raw')
    fls = list(path.rglob('*.parquet'))
    return fls
```

```python
data_files = find_files()
```

```python
schedule = {x.name: pd.read_parquet(x) for x in data_files if  '_schedule' in x.as_posix()}
```

```python
def parse_schedule(file_list):
    df = (
        pd.concat([
            pd.read_parquet(fl) for fl in file_list
            if "_schedule" in fl.as_posix()
        ])
        .rename(columns={"RoundNumber": 'round_number'})
        .set_index(["year", "round_number"])
        .sort_index()
    )
    return df
```

```python
df = parse_schedule(file_list=data_files)
path = Path("../data_parsed/schedule.parquet")
path.parent.mkdir(parents=True, exist_ok=True)
df.to_parquet(path)
```

```python
def parse_drivers(file_list):
    df = (
        pd.concat([
            pd.read_parquet(fl) for fl in file_list
            if "_drivers.parquet" in fl.as_posix()
        ])
        .rename(columns={"index": 'ShortName', "DriverNumber": 'driver_number'})
        .assign(driver_number=lambda df: df['driver_number'].astype(int))
        .set_index(["year", "round_number", 'session_type', 'driver_number'])
        .sort_index()
    )
    return df
```

```python
df = parse_drivers(data_files)
```

```python
def parse_laps(file_list):
    df = (
        pd.concat([
            pd.read_parquet(fl) for fl in file_list
            if "_laps.parquet" in fl.as_posix()
        ])
        .rename(columns={"DriverNumber": 'driver_number', "LapNumber": "lap_number"})
        .assign(driver_number=lambda df: df['driver_number'].astype(int))
        .assign(lap_number=lambda df: df['lap_number'].astype(int))
        .drop('index', axis=1)
    )
    index = ["year", "round_number", 'session_type', 'driver_number', 'lap_number']
    columns = [x for x in df.columns if x not in index]
    df = df[index + columns]
    df = df.sort_values(by=index)
    return df, index
```

```python
df, index= parse_laps(data_files)
```
